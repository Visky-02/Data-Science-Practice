```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
House_prices=pd.read_csv('house_prices.csv',index_col='ID')
```

```python
House_prices
```

```python
House_prices.rename(columns={'Property #':'Property','Area (ft.)':'Area'},inplace=True)
House_prices
```

```python
plt.scatter(House_prices['Area'],House_prices['Price'],
            c=House_prices['Building Type'],
            edgecolor='midnightblue',
            s=80,alpha=0.6,
            cmap='twilight')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.title('House Price Distribution',fontsize=14)
plt.xlabel('Prices',fontsize=10)
plt.ylabel('No of Houses',fontsize=10)
plt.hist(House_prices['Price'],edgecolor='red')
plt.show()
```
