```python
import pandas as pd
udf=pd.read_csv('Universities.csv',index_col='Univ')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(udf)
scaled_data
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
model.fit(scaled_data)
```

```python
m1=model.inertia_
```

```python
m1
```

```python
model.labels_
```

```python
udf['Label']=model.labels_
```

```python
udf
```

```python
grp=udf.groupby('Label')
for x,y in grp:
    print(y)
    print('_'*70)
```

```python
grp.agg('mean')
```
