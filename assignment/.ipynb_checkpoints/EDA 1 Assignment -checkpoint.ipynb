```python
columns_to_clean=['LB','Width','FM','AC']
for col in columns_to_clean:
    q1=cardio[col].quantile(0.25)
    q3=cardio[col].quantile(0.75)
    IQR=q3-q1
    
    lower_bound=q1-1.5*IQR
    upper_bound=q3+1.5*IQR
    cardio[col]=cardio[col].clip(lower=lower_bound,upper=upper_bound)
    
```

```python
cardio.boxplot()
plt.show()
```

Outlier Detection and Treatment:

Detection: Utilized boxplot() to visually identify outliers across variables.

Treatment Strategy: Applied the IQR (Interquartile Range) Capping method exclusively to mechanical/sensor features (LB, AC, FM, Width) to handle machine errors without data loss.

Domain Consideration: Deliberately avoided treating outliers in pathology-related variables (e.g., ASTV, ALTV, MLTV, Decelerations) as extreme values in these columns represent critical fetal distress signals (Pathologic class) and are essential for predictive modeling.

 Findings from the Statistical Summary:

Finding 1 (High Skewness in Pathology Features): Variables representing abnormalities, such as ALTV (Abnormal Long Term Variability), show a massive gap between the Mean (~10.35) and the Median (0.0), coupled with a high standard deviation (21.28). A median of 0 indicates that more than half of the patients have absolutely no abnormal variability (healthy majority). However, the high mean implies an extreme right-skewed distribution, driven by a small subset of critical cases with severe abnormalities.

Finding 2 (Fetal Movement Concentration): The FM (Fetal Movement) feature has an IQR near zero and a Median of 0.0. This implies that for the vast majority of the recording periods, fetal movement detected by the monitor was virtually nonexistent or minimal, with only rare, sudden peaks occurring (Max = 0.98).

Finding 3 (Target Variable Imbalance): The target variable NSP (Normal=1, Suspect=2, Pathologic=3) has a Mean of approximately 1.3 and a Median of 1.0. A mean so close to 1 suggests that the dataset is highly imbalanced, with the overwhelming majority of the records belonging to healthy (Normal) fetuses.

### Categorical Variables Analysis & Inconsistency Resolution:

Justification for Categorical Variable Selection

Data Profiling: Out of the 21 features in this dataset, 19 are continuous numerical measurements (e.g., Fetal Heart Rate, Histogram Widths, Variability Percentages). Creating frequency bar charts or pie charts for continuous variables is analytically incorrect and visually uninterpretable.

Identifying Categories: Upon inspecting the unique values and the dataset's data dictionary, only two variables represent distinct categorical classes (buckets):

NSP: Target variable representing Fetal Health (1 = Normal, 2 = Suspect, 3 = Pathologic).

Tendency: Histogram tendency (-1 = Left asymmetric, 0 = Symmetric, 1 = Right asymmetric).

Data Correction Applied: Due to floating-point anomalies (e.g., storing categorical '2.0' as '1.990484'), a programmatic check failed initially. I explicitly rounded and cast these two variables to integers to restore their categorical integrity before plotting.



Visual Findings: The bar charts confirm that both categorical variables exhibit extreme class imbalance. The vast majority of Fetal Heart Rate patterns show a Tendency of 0 and an NSP class of 1 (Normal).

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.violinplot(x='NSP', y='ASTV', data=cardio, palette='muted')
plt.title('ASTV Distribution across Fetal Health (NSP)')
plt.xlabel('NSP Class (1=Normal, 2=Suspect, 3=Pathologic)')
plt.ylabel('ASTV (Abnormal Short Term Variability %)')

plt.subplot(1, 2, 2)
sns.violinplot(x='NSP', y='ALTV', data=cardio, palette='muted')
plt.title('ALTV Distribution across Fetal Health (NSP)')
plt.xlabel('NSP Class (1=Normal, 2=Suspect, 3=Pathologic)')
plt.ylabel('ALTV (Abnormal Long Term Variability %)')

plt.tight_layout()
plt.show()
```

Advanced Visualization Insights (Violin Plots):

Design Choice: Given the high dimensionality of the dataset (21 features), a full pair plot would result in over 400 subplots, reducing interpretability. Therefore, Violin Plots were selected to perform targeted, deeper analysis on key pathological features (ASTV and ALTV) against the target variable (NSP).

Insight on ASTV & ALTV: The violin plots demonstrate a clear structural shift. For Normal cases (Class 1), the variability is tightly packed near the lower end. However, as fetal health deteriorates to Pathologic (Class 3), the "violins" stretch upwards and become wider at higher values.

Conclusion: This visually confirms that an increase in both Abnormal Short-Term and Long-Term Variability is a strong, distinct indicator of fetal distress.

## Pattern Recognition and Insights

1. Correlations Between Variables and Their Implications:
Based on the correlation heatmap generated in the previous section, several key relationships stand out regarding the target variable NSP (Fetal Health):

Strong Positive Correlations (The Danger Signs): Features like ASTV (Abnormal Short Term Variability), ALTV (Abnormal Long Term Variability), and DP (Prolonged Decelerations) show a positive correlation with NSP.

Implication: As these abnormal variabilities and decelerations increase, the class shifts from 1 (Normal) towards 3 (Pathologic). These are primary indicators of fetal distress and oxygen deprivation.

Strong Negative Correlations (The Healthy Signs): The feature AC (Accelerations) shows a negative correlation with NSP.

Implication: Higher frequency of heart rate accelerations is strongly associated with Class 1 (Normal). Accelerations are a sign of a healthy, responsive nervous system in the fetus.

## Conclusion & Final Report

Final Conclusion and Brief Report
1. Summary of Key Insights and Patterns Discovered:
Through a comprehensive Exploratory Data Analysis (EDA) of the Cardiotocographic dataset, several critical patterns were uncovered:

Data Quality & Integrity: The dataset was largely robust, requiring the removal of only ~1% of rows due to missing values across critical features. A hidden data type inconsistency was resolved by converting floating-point categories back to integers (e.g., NSP, Tendency).

Severe Class Imbalance: The target variable NSP (Fetal Health) is highly skewed. Over 75% of the data falls under Class 1 (Normal), making the 'Suspect' and 'Pathologic' classes a severe minority.

The Danger Metrics (Pathological Indicators): Features such as ASTV (Abnormal Short Term Variability), ALTV (Abnormal Long Term Variability), and DP (Prolonged Decelerations) display a strong positive correlation with fetal distress. The distributions for these features are heavily right-skewed, meaning sudden high spikes are definitive red flags for pathology.

The Health Metrics (Positive Indicators): AC (Accelerations) showed the strongest negative correlation with the pathologic class, validating it as a reliable clinical sign of a healthy, responsive fetus.


### 2. Impact on Decision-Making and Further Analyses:


These insights directly dictate the roadmap for clinical decision-making and future Machine Learning pipelines:

Handling Imbalance in ML: Any future predictive model trained on this raw data will be heavily biased towards predicting 'Normal' cases. Further analysis must implement resampling techniques, such as SMOTE (Synthetic Minority Over-sampling Technique), or utilize class-weighted algorithms to ensure critical 'Pathologic' cases are not missed.

Model Selection: Given the extreme right-skewness and the preservation of critical medical outliers in variables like ASTV and ALTV, non-linear and tree-based models (e.g., Random Forest, XGBoost) should be preferred over linear models like Logistic Regression.

Clinical Decision Support: From a medical standpoint, hospital monitoring systems should be configured to trigger immediate alerts based on spikes in ASTV and DP, rather than relying heavily on the baseline heart rate (LB), which showed a surprisingly weak correlation with the final pathologic outcome.
