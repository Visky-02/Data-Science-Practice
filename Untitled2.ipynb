```python
# stats.norm.interval(confidence, loc=mean, scale=standard_error)
# standard_error = pop_std / sqrt(n)
import numpy as np
from scipy import stats

n = 140
mean = 1990
pop_std = 2500
std_error = pop_std / np.sqrt(n)

z_range = stats.norm.interval(0.95, loc=mean, scale=std_error)
print("Z-Interval:", z_range)
```

### finding value of z 

```python
import scipy.stats as stats
# 95% confidence ke liye z-value
z_value = stats.norm.ppf((1 + 0.95) / 2) 
print(z_value) # Output: 1.9599...
```

# finding value of t

```python
# 15 degree of freedom, 95% confidence
t_value = stats.t.ppf((1 + 0.95) / 2, df=15)
print(t_value) # Output: 2.131...
```
