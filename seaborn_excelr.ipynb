```python
tips=sns.load_dataset('tips')
```

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='box',hue='sex',
           height=3,aspect=1.5,palette='vlag')
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='strip',
           hue='day',height=3,aspect=1.5)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='swarm',
           hue='sex',height=3,aspect=1.5)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='violin',
           hue='day',height=3,aspect=2.5)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='violin',
           hue='sex',height=3,aspect=1.5,split=True)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='point',
           hue='sex',height=3,aspect=1.5)
plt.show()
```

```python
house_prices=pd.read_csv('house_prices.csv')
house_prices
```

```python
house_prices.rename(columns={
    'Area (ft.) ': 'Area'
},inplace=True)
```

```python
house_prices
```

```python
house_prices.rename(columns={
    'Property #': 'Property',
    'Area (ft.)' : 'Area'
}, inplace=True)
```

```python
house_prices
```

```python
sns.relplot(house_prices,x='Area',y='Price',kind='scatter',height=3,aspect=3.5,s=60,hue='Status',palette='twilight')
plt.show()
```

```python
sns.displot(house_prices,x='Price',kind='hist',height=3,aspect=1.5,color='orange',hue='Status')
plt.title('Price Distribution')
plt.show()
```

```python
sns.displot(house_prices,x='Price',kind='hist',height=3,aspect=1.5,color='orange',hue='Status',kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
sns.displot(house_prices,x='Price',kind='kde',height=3,aspect=1.5,color='orange',hue='Status',fill=True)
plt.title('Price Distribution')
plt.show()
```
