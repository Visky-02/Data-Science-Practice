```python
import pandas as pd
import numpy as np
from scipy import stats

# 1. Z-SCORE CALCULATION (From your Image 5)
marks = [43, 55, 67, 75, 60]
mean_marks = np.mean(marks) # 60.0
std_marks = np.std(marks, ddof=0) # 10.8 (Using ddof=0 to match your notebook exact calculation)

# Z-score formula: (X - mean) / std
z_scores = [(x - mean_marks) / std_marks for x in marks]

print("--- Z-SCORES ---")
for mark, z in zip(marks, z_scores):
    print(f"Marks: {mark} | Z-Score: {round(z, 2)}")


# 2. CONFIDENCE INTERVAL (95%) CALCULATION
print("\n--- 95% CONFIDENCE INTERVAL ---")
# Maan le 30 baccho ki average typing speed 45 hai, std 5 hai.
sample_mean = 45 
sample_std = 5
n = 30

# 95% Confidence Level ke liye Z-value approx 1.96 hoti hai
confidence_level = 0.95
# scipy se exact z-value nikalna:
z_value = stats.norm.ppf((1 + confidence_level) / 2) 

# Margin of Error = Z * (std / sqrt(n))
margin_of_error = z_value * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Point Estimate: {sample_mean} WPM")
print(f"Z-Value for 95% Confidence: {round(z_value, 2)}")
print(f"Margin of Error: ±{round(margin_of_error, 2)} WPM")
print(f"Range Estimate (Confidence Interval): {round(lower_bound, 2)} to {round(upper_bound, 2)} WPM")
```
