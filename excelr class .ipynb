### DBSCAN 

```python
import pandas as pd
import numpy as np
iris=pd.read_csv('iris.csv')
```

```python
iris
```

```python
data=iris.drop('species',axis=1)
```

```python
data
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=0.5,min_samples=12)
model.fit(data)
```

```python
model.labels_
```

```python
data['labels']=model.labels_
data.head()
```

```python
grp=data.groupby('labels')
grp.agg('mean')
```

```python
grp2=iris.groupby('species')
grp2.agg('mean')
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data.drop('labels',axis=1),data['labels'])
```
