```python
def check_grade(marks):
    marks=int(input('enter your marks: '))
    if marks>100 and marks<0:
        print('enter the marks between 0-100')
    elif marks>=80:
        print('A')
        return
    elif marks>=50 and marks<80:
        print('B')
        return
    else:
        return 'fail'
    
```

```python
prices = [50, 100, 80, 120, 60]
total = 0

# Ek-ek price list mein se nikalo
for x in prices:
    
    total += x  # Step 1: Pehle current price 'x' ko total mein jodo
    
    # Step 2: Jodne ke baad check karo ki kya total 300 cross kar gaya?
    if total > 300:
        print(f"Total {total} ho gaya hai. 300 ki limit cross! Stopping here.")
        break  # Step 3: Jaise hi > 300 ho, Emergency Brake laga do aur loop tod do

# Ye loop ke bahar hai, final answer batane ke liye
print(f"Final bill amount is: {total}")
```

```python
def assign_grade(speed):
    if speed>=40:
        return "Grade A"
    elif speed>=30 and speed<40:
        return "Grade B"
    else:
        return "Grade c"
    
        
```

```python
student_speeds = [45, 28, 35, 52, 19]
for speed in student_speeds:
    result=assign_grade(speed)
    print("Results of Students are:", result)
    
```

```python
def generate_batch_report(batch_list):
    report={}
    for speed in batch_list:
        if speed>=40:
            grade= "Grade A"
        elif speed>=30:
            grade= "Grade B"
        else:
            grade= "Grade C"
        report[speed]=grade
    return report
student_speeds = [45, 28, 35, 52, 19]
print(generate_batch_report(student_speeds))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm # Normal Distribution ka VIP library

# Maan le tere institute ke 1000 students ka data hai
# Mean speed = 35 WPM, Standard Deviation = 5 WPM
speeds = np.random.normal(loc=35, scale=5, size=1000)
```

```python
speeds
```

```python
sns.histplot(speeds, kde=True)
```

```python
# norm.cdf(40, loc=35, scale=5) batata hai 40 ya usse KAM speed walo ki probability.
# Humein 40 se ZYADA chahiye, toh total (1) mein se minus kar denge.

prob_greater_than_40 = 1 - norm.cdf(40, loc=35, scale=5)
print("Probability of speed > 40 WPM is:", round(prob_greater_than_40 * 100, 2), "%")
```
