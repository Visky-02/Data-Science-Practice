```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train_scaled=scale.fit_transform(X_train)
X_test_scaled=scale.transform(X_test)
```

## Bagging Model

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
lreg=LogisticRegression(max_iter=200)
bagg_model=BaggingClassifier(estimator=lreg,n_estimators=100)
bagg_model.fit(X_train_scaled,y_train)
predi=bagg_model.predict(X_test_scaled)
print('the accuracy is : ',accuracy_score(y_test,predi))
print('the classification report is : ',classification_report(y_test,predi))
```

### RandomForest Classifier

```python
rf_model=RandomForestClassifier(n_estimators=100,max_depth=6)
rf_model.fit(X_train_scaled,y_train)
prd=rf_model.predict(X_test_scaled)
print('Accuracy score of randomforest is : ',accuracy_score(y_test,prd))
print('classification report of randomforest is : ',classification_report(y_test,prd))
```

### Tune Hyperparameter

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'max_depth':[2,4,6,8,10,12],
    'n_estimators':['50','100','200','300']
}
```
