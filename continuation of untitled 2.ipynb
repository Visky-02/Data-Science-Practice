### z test

```python
import numpy as np
from scipy import stats
```

```python
mean=800
std=20
n=50
b=812
```

```python
z=(b-mean)/(std/np.sqrt(50))
```

```python
z
```

```python
stats.norm.cdf(z)
```

```python
p_val=np.round(1-stats.norm.cdf(z))
```

```python
alpha=0.05
if p_val>alpha:
    print('Failed to reject null hypothesis. the light bulb dont last longer than 800 hours')
else:
    print('Failed to reject null hypothesis. the light bulb  last longer than 800 hours')
```

```python
mean_pop=6
std_pop=1.2
n=30
sample_mean=5.56
```

```python
z=(sample_mean-mean_pop)/(std_pop/np.sqrt(30))
```

```python
z
```

```python
p_val=1-stats.norm.cdf(z)
```

```python
p_val
```

```python
p_val1=2*p_val
```

```python
p_val1
```

```python
alpha=0.05
if p_val1<alpha:
    print('reject null hypothesis')
else:
    print('accept alternate hypothesis')
```
