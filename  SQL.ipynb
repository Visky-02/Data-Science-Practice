```python
import pandas as pd
import sqlite3

# 1. Tera Dummy Data (Pandas mein)
data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Amit', 'Priya', 'Rahul', 'Neha', 'Vikas'],
    'Course': ['Steno', 'Typing', 'Steno', 'Steno', 'Typing'],
    'Typing_Speed': [35, 55, 45, 60, 42]
}
df_students = pd.DataFrame(data)

# 2. Jupyter ke andar SQL Database banana (Cheat Code)
conn = sqlite3.connect(':memory:') # RAM mein temporary database banaya
df_students.to_sql('Students_Data', conn, index=False) # Pandas data ko SQL table mein daal diya

# 3. TERA SQL QUERY (Triple quotes mein likhte hain Python mein)
query = """
SELECT Name, Typing_Speed
FROM Students_Data
WHERE Course = 'Steno' AND Typing_Speed > 40
ORDER BY Typing_Speed DESC;
"""

# 4. Result dekhna
result = pd.read_sql(query, conn)
print(result)
```

```python
import pandas as pd
from pandasql import sqldf

# Creating the Students Dataframe
data_students = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Amit', 'Priya', 'Rahul', 'Neha', 'Vikas'],
    'Course': ['Steno', 'Typing', 'Steno', 'Typing', 'Typing'],
    'Age': [19, 20, 20, 21, 19],
    'Typing_Speed': [35, 55, 40, 60, 45]
}
df_students = pd.DataFrame(data_students)

# Creating the Fees Dataframe
data_fees = {
    'Student_ID': [101, 102, 103, 104, 106],
    'Amount_Paid': [5000, 2500, 2000, 5000, 3000],
    'Status': ['Full', 'Half', 'Half', 'Full', 'Half']
}
df_fees = pd.DataFrame(data_fees)
```

```python
b="""
SELECT Name, Typing_Speed,
    CASE
        WHEN Typing_Speed>50 THEN 'Fast'
        WHEN Typing_Speed>=40 THEN 'Average'
        ELSE 'Slow'
    END AS Speed_Status
FROM df_students
"""
sqldf(b)
```

```python
b1="""
WITH Full_Fees_CTE AS(
    SELECT Student_ID,Amount_Paid,Status
    FROM df_fees
    WHERE Status='Full'
)
SELECT * FROM Full_Fees_CTE
"""
sqldf(b1)
```

```python
b_final = """
-- Step 1: Create the Temporary CTE
WITH Full_Fees_CTE AS (
    SELECT Student_ID, Amount_Paid
    FROM df_fees
    WHERE Status = 'Full'
)

-- Step 2: Immediately use it in the Main Query
SELECT df_students.Name, Full_Fees_CTE.Amount_Paid
FROM df_students
JOIN Full_Fees_CTE 
ON df_students.Student_ID = Full_Fees_CTE.Student_ID
"""

sqldf(b_final)
```

```python
c = """
SELECT Name, Course, Typing_Speed,
       ROW_NUMBER() OVER (PARTITION BY Course ORDER BY Typing_Speed DESC) AS Speed_Rank
FROM df_students
"""
sqldf(c)
```

```python
d = """
-- Step 1: CTE ke andar Window Function laga kar sabko rank do
WITH Ranked_Students AS (
    SELECT Name, Course, Typing_Speed,
           ROW_NUMBER() OVER (PARTITION BY Course ORDER BY Typing_Speed DESC) AS Number_Sticker
    FROM df_students
)

-- Step 2: Ab is CTE (Virtual Table) mein se sirf unko nikal jinka Sticker 1 hai
SELECT Name, Course, Typing_Speed
FROM Ranked_Students
WHERE Number_Sticker = 1;
"""
sqldf(d)
```

# Dense Rank Syntax of SQL

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
from pandasql import sqldf

jti_data = {
    'Student_Name': ['Rahul', 'Priya', 'Amit', 'Neha', 'Karan'],
    'Course': ['Stenography', 'DCA', 'Tally Prime', 'Stenography', 'DCA'],
    'Typing_Speed_WPM': [45, 30, 0, 55, 35],'Fees_Paid': [5000, 2500, 3000, 5000, 2000]
}
df11=pd.DataFrame(jti_data)
```

###  Kal raat ka nanga sach yaad kar. Tujhe SQL mein saare baccho ka data dikhana hai, aur unko unki 'Fees_Paid' ke hisaab se RANK karna hai (Sabse zyada fees dene wala Rank 1). Same fees walo ko same rank milna chahiye, aur koi number skip nahi hona chahiye.
Exact SQL query bata. (Hint: OVER(ORDER BY ... DESC)).

```python
q5="""
SELECT Student_Name, Fees_Paid, DENSE_RANK() OVER(ORDER BY Fees_Paid DESC) AS Rank 
FROM df11;
""" 
sqldf(q5)
```
