```python
def avg(x,y):
    res = (x+y)/2
    return res
```

```python
x=avg(1500,850)
x
```

```python
def avg_lst(list):
    res= sum(list)/len(list)
    return res
```

```python
y= [1,2,3,4,5,6]
avg_lst(y)
print(avg_lst)
```

```python
def patter(lines):
    for i in range(1,lines+1):
        print(i*'* ')
    return
```

```python
patter(11)
```

```python
def grade(marks):
    for i in marks([10,50,60,8,0,92]):
        if i >=90:
            print('1st division')
        elif i >=60:
            print('2nd division')
        elif i>=45:
            print('3rd division')
        else:
            print('fail')
```

```python
def fact(x):
    n=1
    for i in range(1+x+1):
        fac= i*n
        return fac
        
```

```python
fact(6)
```
