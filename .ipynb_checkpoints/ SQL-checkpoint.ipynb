```python
import pandas as pd
import sqlite3

# 1. Tera Dummy Data (Pandas mein)
data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Amit', 'Priya', 'Rahul', 'Neha', 'Vikas'],
    'Course': ['Steno', 'Typing', 'Steno', 'Steno', 'Typing'],
    'Typing_Speed': [35, 55, 45, 60, 42]
}
df_students = pd.DataFrame(data)

# 2. Jupyter ke andar SQL Database banana (Cheat Code)
conn = sqlite3.connect(':memory:') # RAM mein temporary database banaya
df_students.to_sql('Students_Data', conn, index=False) # Pandas data ko SQL table mein daal diya

# 3. TERA SQL QUERY (Triple quotes mein likhte hain Python mein)
query = """
SELECT Name, Typing_Speed
FROM Students_Data
WHERE Course = 'Steno' AND Typing_Speed > 40
ORDER BY Typing_Speed DESC;
"""

# 4. Result dekhna
result = pd.read_sql(query, conn)
print(result)
```

```python
from pandasql import sqldf

# Bas apna SQL likh (Dataframe ka naam hi table ka naam ban jayega)
query = """
SELECT Name, Typing_Speed
FROM df_students
WHERE Course = 'Typing' AND Typing_Speed < 50
"""

# Seedha Run
sqldf(query)
```

```python
queryy="""
SELECT Name, Course, Typing_Speed
FROM df_students
WHERE Course='Typing' AND Typing_Speed<50
ORDER BY Typing_Speed ASC
"""
sqldf(queryy)
```

```python
query2="""
SELECT Name,Course,Typing_Speed
FROM df_students
WHERE Course='Typing' and Typing_Speed<50
ORDER BY Typing_Speed ASC;
"""
sqldf(query2)
```

```python
query3 = """
SELECT Course, AVG(Typing_Speed) AS Average_Speed
FROM df_students
GROUP BY Course;
"""

sqldf(query3)
```

```python
query4="""
SELECT Name,Course,Typing_Speed,Typing_Speed+15 AS Target_Speed
FROM df_students
WHERE Course='Steno'
"""
sqldf(query4)
```
