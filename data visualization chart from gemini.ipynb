```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, norm

np.random.seed(42)
df_viz = pd.DataFrame({
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.normal(50000, 15000, 100),
    'Experience': np.random.randint(1, 20, 100),
    'Department': np.random.choice(['IT', 'HR', 'Sales', 'Marketing'], 100),
    'Performance_Score': np.random.uniform(1, 10, 100),
    'Attrition': np.random.choice([0, 1], 100, p=[0.8, 0.2])
})

plt.figure(figsize=(8,4))
sns.histplot(data=df_viz, x='Salary', kde=True)
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(data=df_viz, x='Department', y='Salary')
plt.show()

plt.figure(figsize=(8,4))
sns.violinplot(data=df_viz, x='Department', y='Performance_Score', hue='Attrition', split=True)
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(data=df_viz, x='Experience', y='Salary', hue='Department', size='Performance_Score')
plt.show()

sns.pairplot(df_viz[['Age', 'Salary', 'Experience', 'Performance_Score']])
plt.show()

plt.figure(figsize=(8,4))
sns.heatmap(df_viz[['Age', 'Salary', 'Experience', 'Performance_Score']].corr(), annot=True, cmap='coolwarm')
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(data=df_viz, x='Department', y='Salary', estimator=np.median)
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(data=df_viz, x='Department', hue='Attrition')
plt.show()

plt.figure(figsize=(8,4))
sns.kdeplot(data=df_viz, x='Salary', hue='Department', fill=True)
plt.show()

plt.figure(figsize=(8,4))
sns.lineplot(data=df_viz, x='Experience', y='Salary', ci=None)
plt.show()
```
